```r
# Draw the heatmap
draw_heatmap <- function(sample_dat, batch_dat, class_dat, min_bound){
  # Merge class & batch
  sidebar <- as.matrix(merge(batch_dat, class_dat, by = "sample_id"))
  rownames(sidebar) <- sidebar[,1]
  sidebar <- sidebar[,-1]
  colnames(sidebar) <- c("Batch", "Type")
  # Set the color
  cr <- colorRampPalette(c("navy", "white","firebrick2"), bias = 0.95, space = "rgb")
    if(min_bound != 0){
        ideal_break <- seq(min_bound, 1, (1 - min_bound)/1000)
         s <-  heatmap3(x = sample_dat, 
           ColSideColors = sidebar,
           scale = "none",
           na.rm = T,
           revC = T,
           col = cr(1000),
           breaks = ideal_break,
           showColDendro = F,
           showRowDendro = T,
           labCol = '', labRow = '',
           Colv = NULL,Rowv = NULL)
    }
  else{
       s <- heatmap3(x = sample_dat, 
           ColSideColors = sidebar,
           scale = "none",
           na.rm = T,
           revC = T,
           col = cr(500),
           showColDendro = F,
           showRowDendro = T,
           labCol = '', labRow = '',
           Rowv = NULL, Colv = "Rowv"
           )
  }
  # Set the file name
  file_name <- paste(substitute(sample_dat), ".png", sep = "")
  # Make the plot
  # png(file_name)
  # s
  # dev.off()
  return(s)
}
```

```r
draw_heatmap(merge_dat, merge_batch, merge_class, 0)
```
